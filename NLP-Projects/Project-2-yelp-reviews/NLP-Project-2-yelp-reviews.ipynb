DATA CLEANING/TEST TRAIN!!!

```python
import string

string.punctuation
```

```python
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
##function to clean the data and remove the stopwords
def message_clean(message):
    Test_punc_removed = [char for char in message if char not in string.punctuation]
    Test_punc_removed_join = ''.join(Test_punc_removed)
    Test_punc_removed_join_clean = [word for word in Test_punc_removed_join.split() if word.lower() not in stopwords.words('english')]

    return Test_punc_removed_join_clean
```

```python
yelp_review_df_clean = yelp_review_merge['text'].apply(message_clean)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(analyzer=message_clean)

## we can substitute the above cell and directly do that process in the following code as well
yelp_countvectorizer = vectorizer.fit_transform(yelp_review_merge['text'])
```

```python
print(vectorizer.get_feature_names_out())
```

```python
print(yelp_countvectorizer.toarray())
```

```python
yelp_countvectorizer.shape
```

```python
from sklearn.naive_bayes import MultinomialNB

NB_classifier = MultinomialNB()
target_label = yelp_review_merge['stars'].values
```

```python
NB_classifier.fit(yelp_countvectorizer, target_label)
```

```python
X = yelp_countvectorizer
y = target_label
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
NB_classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
prediction = NB_classifier.predict(X_test)
```

```python
cm = confusion_matrix(y_test, prediction)
```

```python
print('Classification Report:\n')
print(classification_report(y_test, prediction))
```

```python
print('Confusion Matrix:\n')
print(confusion_matrix(y_test, prediction))
sns.heatmap(cm, annot=True)
```
